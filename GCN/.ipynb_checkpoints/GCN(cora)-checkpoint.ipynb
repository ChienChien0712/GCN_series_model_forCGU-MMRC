```python
import math

import torch

from torch.nn.parameter import Parameter
from torch.nn.modules.module import Module

import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.rcParams['figure.dpi'] = 100
```

```python
def encode_onehot(labels):
    classes = set(labels)
    classes_dict = {c: np.identity(len(classes))[i, :] for i, c in
                    enumerate(classes)}
    labels_onehot = np.array(list(map(classes_dict.get, labels)),
                             dtype=np.int32)
    return labels_onehot


def load_data(path="../data/cora/", dataset="cora"):
    """Load citation network dataset (cora only for now)"""
    print('Loading {} dataset...'.format(dataset))
    
    #np.genfromtxt:讀取每一列，製作成二維np.array
    idx_features_labels = np.genfromtxt("{}{}.content".format(path, dataset),dtype=np.dtype(str))
    
    #sp.csr_matrix:Compressed Sparse Row matrix，為壓縮稀疏矩陣的儲存方式
    features = sp.csr_matrix(idx_features_labels[:, 1:-1], dtype=np.float32)
    #encode_onehot:把類別轉成one-hot格式
    labels = encode_onehot(idx_features_labels[:, -1])
    classes = set(idx_features_labels[:, -1])
    classes_dict = {i: c for i, c in enumerate(classes)}    
    ID = idx_features_labels[:, 0]
    # build graph
    #為論文編號建立一個從0開始的index編號
    idx = np.array(idx_features_labels[:, 0], dtype=np.int32)
    idx_map = {j: i for i, j in enumerate(idx)}
    
    #引入兩入文編號之間有連結的資訊
    edges_unordered = np.genfromtxt("{}{}.cites".format(path, dataset),dtype=np.int32)
    #改成index編號的連結
    edges = np.array(list(map(idx_map.get, edges_unordered.flatten())),
                     dtype=np.int32).reshape(edges_unordered.shape)
    #coo_matrix:Coordinate Format，如果很多0，把資料轉成coordinate(座標）的儲存方式，會省很多空間
    #資料(有連結就是1):np.ones(edges.shape[0])；row:edges[:, 0]；column:edges[:, 1]，但這樣只會做出單向連結
    adj = sp.coo_matrix((np.ones(edges.shape[0]), (edges[:, 0], edges[:, 1])),
                        shape=(labels.shape[0], labels.shape[0]),
                        dtype=np.float32)
    #print(adj.toarray())即可看coo_matrix的內容
    
    
    # build symmetric adjacency matrix，製作對稱A矩陣
    #coo_matrix.multiply(another_matrix):Point-wise multiplication by another matrix
    #做完加減乘除後，變成Compressed Sparse Row format(csr matrix)格式
    adj = adj + adj.T.multiply(adj.T > adj) - adj.multiply(adj.T > adj)

    #features做一般的標準化
    features = normalize(features)
    
    #sp.eye():diag(1,1,1,1,...)+一般的標準化
    """compute D^-1 * (A+I)"""
    normalized_adj = normalize(adj + sp.eye(adj.shape[0]))
    
    
    #或者對稱標準化(Laplacian)
    """compute D^-0.5 * (A+I) * D^-0.5"""
    laplacian_adj = laplacian(adj + sp.eye(adj.shape[0]))
    
    
    idx_train = range(140)
    idx_val = range(200, 500)
    idx_test = range(500, 1500)

    features = torch.FloatTensor(np.array(features.todense()))
    #labels原本是矩陣，現在只要找出每列的1的位置。
    labels = torch.LongTensor(np.where(labels)[1])
    normalized_adj = sparse_mx_to_torch_sparse_tensor(normalized_adj)
    laplacian_adj = sparse_mx_to_torch_sparse_tensor(laplacian_adj)
    
    idx_train = torch.LongTensor(idx_train)
    idx_val = torch.LongTensor(idx_val)
    idx_test = torch.LongTensor(idx_test)
    
    #fetures: X矩陣(Float tensor)(32bits)
    #normalized_adj: A矩陣(torch.sparse.FloatTensor (coo))
    #labels: 答案Y矩陣(Long tensor)(64bits)
    return normalized_adj, laplacian_adj, features, labels, idx_train, idx_val, idx_test, ID, classes_dict


def normalize(mx):
    """Row-normalize sparse matrix"""
    rowsum = np.array(mx.sum(1))
    r_inv = np.power(rowsum, -1).flatten()
    r_inv[np.isinf(r_inv)] = 0.
    r_mat_inv = sp.diags(r_inv)
    mx = r_mat_inv.dot(mx)
    return mx

def laplacian(mx):
    """compute L=D^-0.5 * (mx) * D^-0.5"""
    degree = np.array(mx.sum(1))
    d_hat = sp.diags(np.power(degree, -0.5).flatten())
    laplacian_mx = d_hat.dot(mx).dot(d_hat)
    return laplacian_mx

def accuracy(output, labels):
    preds = output.max(1)[1].type_as(labels)
    correct = preds.eq(labels).double()
    correct = correct.sum()
    return correct / len(labels)


def sparse_mx_to_torch_sparse_tensor(sparse_mx):
    """Convert a scipy sparse matrix to a torch sparse tensor."""
    #csr matrix轉回coo matrix
    sparse_mx = sparse_mx.tocoo().astype(np.float32)
    #indices:儲存row和col
    indices = torch.from_numpy(
        np.vstack((sparse_mx.row, sparse_mx.col)).astype(np.int64))
    #values:儲存非0數值
    values = torch.from_numpy(sparse_mx.data)
    shape = torch.Size(sparse_mx.shape)
    return torch.sparse.FloatTensor(indices, values, shape)
```

```python
normalized_adj, laplacian_adj, features, labels, idx_train, idx_val, idx_test, ID, classes_dict = load_data(path="./cora/")
```

```python
import torch.nn as nn

class GraphConvolution(Module):
    """
    Simple GCN layer, similar to https://arxiv.org/abs/1609.02907
    """

    def __init__(self, in_features, out_features, bias=True):
        super(GraphConvolution, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.weight = Parameter(torch.FloatTensor(in_features, out_features))
        if bias:
            self.bias = Parameter(torch.FloatTensor(out_features))
        else:
            self.register_parameter('bias', None)
        self.reset_parameters()

    def reset_parameters(self):
        stdv = 1. / math.sqrt(self.weight.size(1))
        self.weight.data.uniform_(-stdv, stdv)
        if self.bias is not None:
            self.bias.data.uniform_(-stdv, stdv)

    def forward(self, input, adj):
        support = torch.mm(input, self.weight)
        output = torch.spmm(adj, support)
        if self.bias is not None:
            return output + self.bias
        else:
            return output

    def __repr__(self):
        return self.__class__.__name__ + ' (' \
               + str(self.in_features) + ' -> ' \
               + str(self.out_features) + ')'
class GCN(nn.Module):
    def __init__(self, nfeat, nhid, nclass, dropout):
        super(GCN, self).__init__()
        #nfeat = 100
        #nhid = '50,20'
        #nclass = 2
        nhid = nhid.split(',')
        nhid = list(map(int,nhid))
        nhid_list = [nfeat] + nhid + [nclass]
        self.nlayer = len(nhid_list)-1
        for i in range(1,len(nhid_list)):
            setattr(self, 'gc%s'%i, GraphConvolution(nhid_list[i-1], nhid_list[i]))
        self.dropout = dropout

    def forward(self, x, adj):
        for i in range(1,self.nlayer): 
            x = F.relu(getattr(self,'gc%s'%i)(x, adj))
            x = F.dropout(x, self.dropout, training=self.training)
        x = getattr(self,'gc%s'%self.nlayer)(x, adj)
        return F.log_softmax(x, dim=1)
```

```python
import time
import argparse #命令列參數
import numpy as np

import torch
import torch.nn.functional as F
import torch.optim as optim
```

```python
#建立命令列參數，之後就可以在command line中輸入並設定參數
parser = argparse.ArgumentParser()
#"--"開頭，代表可選參數(optional argument)，可在command line以[--seed 數值]來設定，若不給定也沒default，則該參數為None。
#action="store_true"代表在可在command line以[--fastmode]來開啟，有輸入開啟為True，沒輸入為False。
#help='文字內容'代表在command line輸入[python3 執行檔案.py --help]，可以看所有參數(Arguments)的幫助訊息，包括positional arguments和optional arguments
parser.add_argument('--no-cuda', action='store_true', default=False,
                    help='Disables CUDA training.')
parser.add_argument('--fastmode', action='store_true', default=False,
                    help='Validate during training pass.')
parser.add_argument('--seed', type=int, default=42, help='Random seed.')
#epochs：每個training data被遍歷的次數
parser.add_argument('--epochs', type=int, default=200,
                    help='Number of epochs to train.')
parser.add_argument('--lr', type=float, default=0.01,
                    help='Initial learning rate.')
parser.add_argument('--weight_decay', type=float, default=5e-4,
                    help='Weight decay (L2 loss on parameters).') # L2 loss
parser.add_argument('--hidden', type=str, default='200,16,8',
                    help='Number of hidden units.')
parser.add_argument('--dropout', type=float, default=0.5,
                    help='Dropout rate (1 - keep probability).')

args = parser.parse_args([]) #以list儲存

#如果torch.cuda.is_available()=True就開啟cude為True
args.cuda = not args.no_cuda and torch.cuda.is_available() 

#設置seed
np.random.seed(args.seed)
torch.manual_seed(args.seed)
if args.cuda:
    torch.cuda.manual_seed(args.seed)
```

```python
model = GCN(nfeat=features.shape[1],
            nhid=args.hidden,
            nclass=len(set(labels.tolist())), #+1是因為原本的label是0,1,2,3,4,5,6，從0開始，所以output的數目應該+1
            dropout=args.dropout)
optimizer = optim.Adam(model.parameters(),lr=args.lr, weight_decay=args.weight_decay)
```

```python
if args.cuda:
    model.cuda()
    features = features.cuda()
    normalized_adj = normalized_adj.cuda()
    labels = labels.cuda()
    idx_train = idx_train.cuda()
    idx_val = idx_val.cuda()
    idx_test = idx_test.cuda()


#設定train一次的過程
Loss_Train = []
Acc_Train = []
Loss_Val = []
Acc_Val = []
def train(epoch):
    t = time.time()
    #把model變成訓練模式，
    model.train()
    optimizer.zero_grad() #清除梯度
    output = model(features, normalized_adj) #就是model.forward(features, normalized_adj)
    # negative log likelihood loss
    #loss定義為：正確位置的log_softmax數值
    loss_train = F.nll_loss(output[idx_train], labels[idx_train]) #運練用idx_train 
    acc_train = accuracy(output[idx_train], labels[idx_train])
    Loss_Train.append(loss_train.tolist())
    Acc_Train.append(acc_train.tolist())
    #反向傳遞計算梯度，取得梯度
    loss_train.backward()
    #更新模型參數
    optimizer.step()

    if not args.fastmode:
        # Evaluate validation set performance separately,
        # deactivates dropout during validation run.
        model.eval()
        output = model(features, normalized_adj)

    loss_val = F.nll_loss(output[idx_val], labels[idx_val]) 
    acc_val = accuracy(output[idx_val], labels[idx_val])
    Loss_Val.append(loss_val.tolist())
    Acc_Val.append(acc_val.tolist())
    print('Epoch: {:04d}'.format(epoch+1),
          'loss_train: {:.4f}'.format(loss_train.item()),
          'acc_train: {:.4f}'.format(acc_train.item()),
          'loss_val: {:.4f}'.format(loss_val.item()),
          'acc_val: {:.4f}'.format(acc_val.item()),
          'time: {:.4f}s'.format(time.time() - t))

prediction = []
def test():
    model.eval()
    output = model(features, normalized_adj) # = model.forward(features, normalized_adj)
    loss_test = F.nll_loss(output[idx_test], labels[idx_test])
    acc_test = accuracy(output[idx_test], labels[idx_test])
    prediction.append(output)
    print("Test set results:",
          "loss= {:.4f}".format(loss_test.item()),
          "accuracy= {:.4f}".format(acc_test.item()))


# Train model
t_total = time.time()
#train epoch次
for epoch in range(args.epochs):
    train(epoch)
print("Optimization Finished!")
print("Total time elapsed: {:.4f}s".format(time.time() - t_total))

# Testing
test()
```

```python
Epoch = range(1,args.epochs+1)
plt.plot(Epoch,Loss_Train,label='training loss')
plt.plot(Epoch,Acc_Train,label='training accuracy')
plt.plot(Epoch,Loss_Val,label='validation loss')
plt.plot(Epoch,Acc_Val,label='validation accuracy')

plt.xlabel('epoch',fontsize=18)
plt.ylabel('loss or accuracy',fontsize=18)
plt.legend(fontsize=18)
plt.grid(linestyle='--')
plt.show()
```
