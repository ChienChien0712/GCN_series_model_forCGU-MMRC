## 建立單層Layer結構

```python
import math
import torch
from torch.nn.parameter import Parameter
from torch.nn.modules.module import Module
```

```python
#建立層(Layer)結構，單繼承自torch.nn.modules.module.Module
class GraphConvolution(Module):
    """
    Simple GCN layer, similar to https://arxiv.org/abs/1609.02907
    """
    def __init__(self, in_features, out_features, bias=True):
        super(GraphConvolution, self).__init__() #繼承父類所有屬性
        #1. X矩陣(特徵,features)的column數，也就是自變量的數目
        self.in_features = in_features
        #2. Output數值的數目。
        self.out_features = out_features
        #3. W矩陣：利用Parameter建立in_features*out_features的矩陣，並且隨機給定起始值
        self.weight = Parameter(torch.FloatTensor(in_features, out_features))
        #4.1. Bias參數矩陣（也是要估計）：建立1*out_features的矩陣，且隨機給定起始值
        if bias:
            self.bias = Parameter(torch.FloatTensor(out_features))
        #4.2. 如果不設定bias，就告訴Module下parameter的'bias'是None
        else:
            self.register_parameter('bias', None)
        #5. 初始化所有參數的起始值
        self.reset_parameters()

    def reset_parameters(self):
        stdv = 1. / math.sqrt(self.weight.size(1)) #size(1)=out_features
        self.weight.data.uniform_(-stdv, stdv) #初始化W矩陣
        if self.bias is not None:
            self.bias.data.uniform_(-stdv, stdv) #初始化Bias矩陣

    #一定要定義forward
    def forward(self, input, adj):
        #1. XW
        support = torch.mm(input, self.weight) #mm(matrix multiplication):兩矩陣相乘
        #2. (標準化A)XW
        output = torch.spmm(adj, support) #spmm(sparse matrix multiplication):兩矩陣相乘
        if self.bias is not None:
        #3.1 (標準化A)XW+Bias
            return output + self.bias
        else:
        #3.2 (標準化A)XW
            return output

    def __repr__(self):
        return self.__class__.__name__ + ' (' \
               + str(self.in_features) + ' -> ' \
               + str(self.out_features) + ')'
```

## 建立GCN結構

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
#由單層GCN，建構出多層的GCN結構。下列範例為2層。
class GCN(nn.Module):
    def __init__(self, nfeat, nhid, nclass, dropout):
        super(GCN, self).__init__() #繼承torch.nn.Module的所有屬性
        self.gc1 = GraphConvolution(nfeat, nhid)
        self.gc2 = GraphConvolution(nhid, nclass)
        self.dropout = dropout

    def forward(self, x, adj):
        #1. (標準化A)XW+Bias
        x = F.relu(self.gc1(x, adj))#調用gc1.forward
        #2. 隱藏神經元：隨機將部分輸出過濾成0，通常dropout rate = 0.5
        x = F.dropout(x, self.dropout, training=self.training) #模型的training狀態參數
        #3. (標準化A)XW+Bias
        x = self.gc2(x, adj)
        return F.log_softmax(x,dim=1)#最終輸出轉log_softmax（機率）
```

## 建立資料處理會用到的函數

```python
import torch
import numpy as np
import scipy.sparse as sp
```

```python
#1. 把類別變項的labels(正確答案)，轉成數字表示的矩陣。
def encode_onehot(labels):
    classes = set(labels)
    classes_dict = {c: np.identity(len(classes))[i, :] for i, c in
                    enumerate(classes)}
    labels_onehot = np.array(list(map(classes_dict.get, labels)),
                             dtype=np.int32)
    return labels_onehot

#2. 一般標準化
def normalize(mx):
    """compute D^-1 * (mx)"""
    rowsum = np.array(mx.sum(1))
    r_inv = np.power(rowsum, -1).flatten()
    r_inv[np.isinf(r_inv)] = 0.
    r_mat_inv = sp.diags(r_inv)
    mx = r_mat_inv.dot(mx)
    return mx

#3. 對稱標準化(Symmetric normalized Laplacian)
def laplacian(mx):
    """compute D^-0.5 * (mx) * D^-0.5"""
    degree = np.array(mx.sum(1))
    d_hat = sp.diags(np.power(degree, -0.5).flatten())
    laplacian_mx = d_hat.dot(mx).dot(d_hat)
    return laplacian_mx

#4. 計算預測結果的正確率
def accuracy(output, labels):
    #預測結果：找每一個row的output(機率)最大值(.max(1))的索引[1]
    preds = output.max(1)[1].type_as(labels) 
    #判斷正確與否：看preds(預測結果)是否等於標籤，並把布林值換成0或1(.double())
    correct = preds.eq(labels).double()
    #計算預測正確的節點數
    correct = correct.sum()
    #回傳正確率
    return correct / len(labels)

#5. 把稀疏矩陣，轉成torch稀疏矩陣
def sparse_mx_to_torch_sparse_tensor(sparse_mx):
    """Convert a scipy sparse matrix to a torch sparse tensor."""
    #csr matrix轉回coo matrix
    sparse_mx = sparse_mx.tocoo().astype(np.float32)
    #indices:儲存row和col
    indices = torch.from_numpy(np.vstack((sparse_mx.row, sparse_mx.col)).astype(np.int64))
    #values:儲存非0數值
    values = torch.from_numpy(sparse_mx.data)
    shape = torch.Size(sparse_mx.shape)
    return torch.sparse.FloatTensor(indices, values, shape)
```

## 建立load data函數

```python
import pandas as pd
import random
```

```python
#6. 引入資料（load data）
def load_data(path="./TestData/", dataset="GCN_TestData.xlsx"):
    """Load citation network dataset"""

    #製作labels矩陣(Y)，製作成二維np.array
    idx_labels = pd.read_excel("./%s/%s"%(path,dataset),sheet_name='Y', header=None)
    #encode_onehot:把類別轉成one-hot格式
    labels = encode_onehot(np.array(idx_labels[0]))
    
    #sp.csr_matrix:Compressed Sparse Row matrix，為壓縮稀疏矩陣的儲存方式
    #製作X矩陣
    idx_feature = pd.read_excel("./%s/%s"%(path,dataset), sheet_name='X', header=None)
    features = sp.csr_matrix(np.array(idx_feature), dtype=np.float32)
    
    #製作A矩陣
    idx_adj = pd.read_excel("./%s/%s"%(path,dataset),sheet_name='A', header=None)
    adj = sp.coo_matrix(np.array(idx_adj),dtype=np.float32)
    

    #features做一般標準化
    features = normalize(features)
    
    #adj做一般標準化
    normalized_adj = normalize(adj) 
    
    #adj做對稱標準化(Laplacian)
    laplacian_adj = laplacian(adj)
    
    #建立四組(train, validation, test, unknown)的index
    gp = set(idx_labels[(idx_labels[0] == 0)|(idx_labels[0] == 1)].index)

    #規劃每組size
    known_size = len(gp)
    train_size = round(known_size*0.8)
    validation_size = round((known_size - train_size)/2)
    test_size = known_size - train_size - validation_size    
    #抽樣
    idx_train = set(random.sample(gp,k=train_size))
    gp = gp - idx_train
    idx_val = set(random.sample(gp,k=validation_size))
    gp = gp - idx_val
    idx_test = set(random.sample(gp,k=test_size))
    #完成建立
    idx_train = sorted(list(idx_train))
    idx_val = sorted(list(idx_val))
    idx_test = sorted(list(idx_test))    
    idx_unknown = list(idx_feature[idx_feature[0] == 2].index)
    
    #轉成Tensor
    features = torch.FloatTensor(np.array(features.todense()))
    #labels原本是矩陣，現在只要找出每列的1的位置。
    labels = torch.LongTensor(np.where(labels)[1])
    normalized_adj = sparse_mx_to_torch_sparse_tensor(normalized_adj)
    laplacian_adj = sparse_mx_to_torch_sparse_tensor(laplacian_adj)
    
    idx_train = torch.LongTensor(idx_train)
    idx_val = torch.LongTensor(idx_val)
    idx_test = torch.LongTensor(idx_test)
    
    #fetures: X矩陣(Float tensor)(32bits)
    #normalized_adj: A矩陣(torch.sparse.FloatTensor (coo))
    #labels: 答案Y矩陣(Long tensor)(64bits)
    return normalized_adj, laplacian_adj, features, labels, idx_train, idx_val, idx_test
```

## 設定參數

```python
import argparse 
import numpy as np
import torch
```

```python
#建立命令列參數，之後就可以在command line中輸入並設定參數
parser = argparse.ArgumentParser()

#"--"開頭，代表可選參數(optional argument)，可在command line以[--seed 數值]來設定，若不給定也沒default，則該參數為None。
#action="store_true"代表在可在command line以[--fastmode]來開啟，有輸入開啟為True，沒輸入為False。
#help='文字內容'代表在command line輸入[python3 執行檔案.py --help]，可以看所有參數(Arguments)的幫助訊息，包括positional arguments和optional arguments
parser.add_argument('--no-cuda', action='store_true', default=False,
                    help='Disables CUDA training.')
parser.add_argument('--fastmode', action='store_true', default=False,
                    help='Validate during training pass.')
parser.add_argument('--seed', type=int, default=42, help='Random seed.')
parser.add_argument('--epochs', type=int, default=200,
                    help='Number of epochs to train.')
parser.add_argument('--lr', type=float, default=0.01,
                    help='Initial learning rate.')
parser.add_argument('--weight_decay', type=float, default=5e-4,
                    help='Weight decay (L2 loss on parameters).')
parser.add_argument('--hidden', type=int, default=8,
                    help='Number of hidden units.')
parser.add_argument('--dropout', type=float, default=0.5,
                    help='Dropout rate (1 - keep probability).')

args = parser.parse_args([]) #以list儲存

#如果torch.cuda.is_available()=True就開啟cude為True
args.cuda = not args.no_cuda and torch.cuda.is_available() 

#設置seed
np.random.seed(args.seed)
torch.manual_seed(args.seed)
if args.cuda:
    torch.cuda.manual_seed(args.seed)
```

```python
normalized_adj, laplacian_adj, features, labels, idx_train, idx_val, idx_test = load_data(path="./TestData/",dataset="GCN_TestData.xlsx")
```

## 建立Model與Optimizer & 訓練model

```python
import torch.optim as optim
import time
import torch.nn.functional as F
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.rcParams['figure.dpi'] = 100
```

```python
#建立GCN模型
model = GCN(nfeat=features.shape[1],
            nhid=args.hidden,
            nclass=labels.max().item() , #因為2是unknown，所以不要+1
            dropout=args.dropout)

#建立自動Gradient Descent的Optimizer，這邊使用Adam法。
optimizer = optim.Adam(model.parameters(),
                       lr=args.lr, weight_decay=args.weight_decay)


#如果可以使用GPU加速，就把資料都轉乘cuda
if args.cuda:
    model.cuda()
    features = features.cuda()
    normalized_adj = normalized_adj.cuda()
    laplacian_adj = laplacian_adj.cuda()
    labels = labels.cuda()
    idx_train = idx_train.cuda()
    idx_val = idx_val.cuda()
    idx_test = idx_test.cuda()


#設定train一次的過程
LossTrain = []
AccTrain = []
LossVal = []
AccVal = []

def train(epoch):
    t = time.time()
    #開啟訓練模式
    model.train()
    #清除梯度，否則梯度會進行累加，梯度累加會用在有batch的訓量當中。
    optimizer.zero_grad()
    #執行model.forward(features, laplacian_adj)，獲得y_hat
    output = model(features, laplacian_adj)
    #print(np.exp(output[idx_train].tolist()))
    #print(output[idx_train].max(1)[1].type_as(labels))
    #自動生成loss function
    #nll_loss: negative log likelihood loss
    loss_train = F.nll_loss(output[idx_train], labels[idx_train]) #運練用idx_train 
    LossTrain.append(loss_train)
    #計算training data的accuracy
    acc_train = accuracy(output[idx_train], labels[idx_train])
    AccTrain.append(acc_train)
    #自動反向傳遞計算各參數的梯度，取得梯度
    loss_train.backward()
    #自動更新模型參數
    optimizer.step()

    #fastmode為False時執行。
    if not args.fastmode:
        #用評價(evaluation)模式，來計算validation data的loss和accuracy
        model.eval()
        output = model(features, laplacian_adj)

    #計算validation data的loss和accuracy
    loss_val = F.nll_loss(output[idx_val], labels[idx_val])
    LossVal.append(loss_val)
    acc_val = accuracy(output[idx_val], labels[idx_val])
    AccVal.append(acc_val)
    print('Epoch: {:04d}'.format(epoch+1),
          'loss_train: {:.4f}'.format(loss_train.item()),
          'acc_train: {:.4f}'.format(acc_train.item()),
          'loss_val: {:.4f}'.format(loss_val.item()),
          'acc_val: {:.4f}'.format(acc_val.item()),
          'time: {:.4f}s'.format(time.time() - t))


def test():
    #在預測testing data時，一定要開啟評價(evaluation)模式
    model.eval()
    #執行model.forward(features, laplacian_adj)，獲得y_hat(預測值)
    output = model(features, laplacian_adj) 
    loss_test = F.nll_loss(output[idx_test], labels[idx_test])
    acc_test = accuracy(output[idx_test], labels[idx_test])
    print("Test set results:",
          "loss= {:.4f}".format(loss_test.item()),
          "accuracy= {:.4f}".format(acc_test.item()))


# Train model
t_total = time.time()
#train epoch次
for epoch in range(args.epochs):
    train(epoch)
print("Optimization Finished!")
print("Total time elapsed: {:.4f}s".format(time.time() - t_total))

# Testing
test()

plt.plot(range(len(LossTrain)),LossTrain, label='Loss_train')
plt.plot(range(len(AccTrain)),AccTrain, label='Accuracy_train')
plt.plot(range(len(LossTrain)),LossTrain, label='Loss_validation')
plt.plot(range(len(AccTrain)),AccTrain, label='Accuracy_validation')

plt.legend(fontsize=12,loc='upper right')
plt.grid(linestyle='--')

plt.show()
```
